```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"

import numpy as np
data = np.genfromtxt(url, delimiter=',', skip_header=1, usecols=1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure (figsize=(10, 6))
plt.plot (data, label="Datos Originales", linestyle='--', marker='x')
plt.title('Pasajeros de aerolíneas internacionales (1949-1996)')
plt.xlabel('Meses desde enero 1949')
plt.ylabel('Número de pasajeros')
plt.grid(True)
plt.legend()
plt.show()
```

```python
log_data = np.log(data)
x = np.arange(1, len(data) + 1)
X = np.column_stack ((np.ones(len(x)),x)) ##colum:stack apila columas para formar matriz
```

```python
##Modelado exponencial, pseudo inversa de x para calcular 2 coeficientes
X_pinv = np.linalg.pinv(X)
params = np.dot(X_pinv, log_data)
print(params)
```

```python
log_A = params[0]
B = params[1]
A = np.exp(log_A)

fitted_values = A * np.exp(B * x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot (data, label = 'Datos Originales', linestyle='--', marker='x')
plt.plot (fitted_values, label='Modelo ajustado', color = 'red')
plt.title('MODELO FINAL MODELO CRECIMIENTO AEROLINEAS')
plt.xlabel('Meses desde enero 1949')
plt.ylabel('Número de pasajeros')
plt.grid(True)
plt.legend()
plt.show()
```
